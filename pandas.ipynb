```python
sorted_runs_2017 = redsox2017[['HR', 'RBI']].sort_values(['HR', 'RBI'], ascending=False).reset_index(drop=True)
sorted_runs_2017
sorted_runs_2017['HR'].describe(include='all')
```

```python
# Excersise #1
# Get total number of home runs and rbi's for the Red Sox
# step 1: Add a new column with the key 'Team' and all column values should be 'BOS'
# step 2: Group by the 'Team' column and get total home runs and rbi's
# Produce data for both 2017 and 2018 (ie print both seperated by a newline character \n
```

```python
runs_17 = redsox2017['HR']
rbis_17 = redsox2017['RBI']
sox17 = runs_17+rbis_17
runs_18 = redsox2018['HR']
rbis_18 = redsox2018['RBI']
sox18 = runs_18+rbis_18
BOS = 'BOS'

data = {
    '2017 HR': sox17,
    '2018 HR': runs_18,
    '2017 RBI': rbis_17,
    '2018 RBI': rbis_18,
    'SOX17': sox17,
    'SOX18': sox18,
    'Team': BOS
}

redsox = pd.DataFrame.from_dict(data)
redsox
```

```python
redsox.groupby('Team', as_index = False).count()[['SOX17', 'SOX18']].sort_values('SOX17', ascending=False)
```
